# concatenar rating


```python
df1 = pd.read_csv('Dataset/ratings/1.csv')
df2 = pd.read_csv('Dataset/ratings/2.csv')
df3 = pd.read_csv('Dataset/ratings/3.csv')
df4 = pd.read_csv('Dataset/ratings/4.csv')
df5 = pd.read_csv('Dataset/ratings/5.csv')
df6 = pd.read_csv('Dataset/ratings/6.csv')
df7 = pd.read_csv('Dataset/ratings/7.csv')
df8 = pd.read_csv('Dataset/ratings/8.csv')
```

```python
frames =[df1,df2,df3,df4,df5,df6,df7,df8]
rating_unido = pd.concat(frames, axis=0)
```

```python
rating_unido.shape
```

```python
#sacamos el promedio de rating agrupado por movieId
score_rating_unido = rating_unido.groupby('movieId')['rating'].mean()
```

```python
score_rating_unido
```

```python
#eliminamos los duplicados
rating_unido = rating_unido.drop_duplicates(subset='movieId')
```

```python
#concatenamos los promedios de rating a score de las plataformas
df_total['score'] = rating_unido.groupby('movieId')['rating'].transform(lambda x: score_rating_unido.loc[x.name])
```
